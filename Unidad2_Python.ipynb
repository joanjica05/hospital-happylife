```python
edad = 21

# mayor que >,
if edad > 30:
  print("Es mayor de edad")

# menor que <,
elif edad < 17:
  print("Es menor de edad")

# mayor igual que >=,
elif edad >= 32:
  print("mayor e igual que 18")

# menor igual que <=,
elif edad <= 2:
  print("Menor e igual que 21")

# igual que ==,
elif edad == 21:
  print("Es igual a 21")

# diferente de !=
elif edad != 21:
  print("No es igual a 21")

# else  = FIN CONDICION -> SI TODO LO ANTERIOR NO SE CUMPLE HAGA LO SIGUIENTE
else:
  print("No se cumplieron condiciones")
```

# ¿Qué es una lista?
Una lista es una colección ordenada, que puede modificar los elementos que contenga.

**Su usan corchetes [ ].**

```python
#Actividad Lista
# recorrer, Ordenar e imprimir de mayor a menor la siguiente lista
#ListaTaller = [1,61,4,2,6]
# sorted = ordenamiento - reverse define ascendente (False) o descendente (True)
lista = [1,2,3]
for lista in sorted(lista, reverse=True):
    print(lista)
```

```python
nombre = ["carlos", "juan", "ana"]
nombre.sort()
print(nombre)
```

# ¿Qué es una tupla?
Una tupla es una colección de datos similar a una lista, pero NO se puede modificar los elementos que contenga.

**Su usan paréntesis ( ).**

```python
# TUPLAS
personas = (5, 4 ,1)
# Comentaarios con numeral #
# print(personas[0])

for persona in sorted(personas, reverse=False):
    print("--")


# Actividad Taller
# recorrer, Ordenar e imprimir de menor a mayor la siguiente tupla
tuplaTaller = ("Kia", "Ford", "Toyota", "Volvo", "Voevo")

for persona in sorted(tuplaTaller, reverse=False):
    print(persona)
```

# ¿Qué es un diccionario?
Un diccionario es como una lista, pero cuenta con **CLAVE** : **VALOR** con o sin orden alguno. Pueden tener  diccionario dentro de otro diccionario.

**Su usan llaves { }.**

```python
# DICCIONARIOS
clientes = {
#    "CLAVE"    : "VALOR",
#     CLAVE     : { CLAVE : VALOR }
    "c01" : {"nombre"  : "Laura","edad"    : 30.5,"ciudad"  : "Bogotá", "pais"    : "Colombia"},
    "c02" : {"nombre"  : "Carlos","edad"   : 25.5,"ciudad"  : "Medellin","pais"  : "Colombia"},
}
#Nombres de las columnas
print("Id\tNombre\tEdad\tCiudad\tPaís")
#recorrido que hace y asigna a cada columna
for id, c in clientes.items():
    print(f"{id} \t{c['nombre']} \t{c['edad']}\t{c['ciudad']}\t{c['pais']}")

# Agregar un nuevo registro
print("--- Agregando un nuevo registro de Perú")
nuevoRegistro = "c03"
clientes[nuevoRegistro] = {"nombre"  : "Demo","edad" : 37, "ciudad" : "Lima", "pais" : "Perú"}

# Listar conjunto de datos con el nuevo registro
print("--- Nuevo registro de Perú")
print("Id\tNombre\tEdad\tCiudad\tPaís")
for id, c in clientes.items():
    print(f"{id} \t{c['nombre']} \t{c['edad']}\t{c['ciudad']}\t{c['pais']}")

# Editar un registro
clientes["c03"]["nombre"] = "Claudia"
clientes["c03"]["edad"] = 25

print("--- Registro C01 actualizado")
print("Id\tNombre\tEdad\tCiudad\tPaís")
for id, c in clientes.items():
    print(f"{id} \t{c['nombre']} \t{c['edad']}\t{c['ciudad']}\t{c['pais']}")

# eliminar un registro
del clientes["c02"]

print("--- Eliminando C02")
print("Id\tNombre\tEdad\tCiudad\tPaís")
for id, c in clientes.items():
    print(f"{id} \t{c['nombre']} \t{c['edad']}\t{c['ciudad']}\t{c['pais']}")
```

```python
#Cree un diccionario de elementos e imprima la lista, los objetos deben ser de su elección
conjunto = {
    "marca":"Toshiba",
    "categoria":"Laptop",
    "procesador":"AMD64",
}
print("Conjunto 1: ",conjunto)
#Agregue 3 nuevos atributos a la lista e imprima
nuevos = {
    "Demo":"Demo 1",
    "Demo1":"Demo 2",
}
#conjunto.update(nuevos)
for clave, valor in nuevos.items():
    conjunto[clave] = valor

print("Conjunto 2: ",conjunto)
#Modifique 2 valores de lo elementos que vea conveniente e imprima el diccionario recorriendolo
conjunto.update({"Demo":"Demo 3","Demo1":"Demo 4"})

for clave, valor in conjunto.items():
    print(f"{clave} : {valor}")
```
